```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
import pickle
```

```python
dataset = "Skills.xlsx"
data = pd.read_excel(dataset)
data.head(5)
```

```python
data_filtered = data[data['Data Value'] >= 5]
```

```python
#pivot table
pivot = data_filtered.pivot_table(index='Title', columns='Element Name', values='Data Value', fill_value=0)
```

```python
#fake 'Personality' and 'Interest' features
pivot['Personality'] = np.random.choice(['Introvert', 'Extrovert', 'Ambivert'], size=len(pivot))
pivot['Interest'] = np.random.choice(['Tech', 'Business', 'Design', 'Science'], size=len(pivot))
```

```python
#categorical features
le_personality = LabelEncoder()
le_interest = LabelEncoder()
le_title = LabelEncoder()
```

```python
pivot['personality_enc'] = le_personality.fit_transform(pivot['Personality'])
pivot['interest_enc'] = le_interest.fit_transform(pivot['Interest'])
pivot['target'] = le_title.fit_transform(pivot.index)
```

```python
#features
X = pivot[['personality_enc', 'interest_enc']]
y = pivot['target']
```

```python
#model Training
model = RandomForestClassifier()
model.fit(X, y)
```

```python
#saving models
with open("career_advisor_model.pkl", "wb") as f:
    pickle.dump(model, f)

with open("le_personality.pkl", "wb") as f:
    pickle.dump(le_personality, f)

with open("le_interest.pkl", "wb") as f:
    pickle.dump(le_interest, f)

with open("le_title.pkl", "wb") as f:
    pickle.dump(le_title, f)
```

```python
#functions to predict careers
def predict_career(personality, interest):
    p = le_personality.transform([personality])[0]
    i = le_interest.transform([interest])[0]
    pred = model.predict([[p, i]])[0]
    return le_title.inverse_transform([pred])[0]
```

```python
#testing the function
print("Recommended career:", predict_career('Introvert', 'Tech'))
```
